# Model Calibration

## What do we mean by model calibration?
- Let's understand this with an example. Let's say we train a model to identify cats and say the model predicts cat with an average probability of 80%. This means that ideally 8 out 10 cat images should be correctly predicted as cat.That is if the avrage confidence is 80% then accuracy as well should be 80 % to call the model as well-calibrated.
- However it has been observed that the modern deep neural networks tends to be over-condfident. The model could be both over confident and undes confident/ such model arecalled as ill-calibrated model.
- There are ways to measure the model calibration, one popular visual way is the relaibility diagram. We will discuss in detail how to plot this kind of reliability diagrams for both classification and regression problem. For now lets just observe a relaibility diagram for a classification problem
- The reliability diagram for a classification problem is shown below
    - Reliability plot provide visual representation of the prediction calibration by plotting the expected sample accuracy as a function of predicted confidence.
    - the ideal curve is the identity curce where the predicted confidence and the observed accuracy (number of samples correctly predicted for that confidence) are same
    - The region above the ideal curve tells that the it is overconfident in this region and the region below tells that the model is over-confident 




## Why do we need model calibration ?
- As more and more neural networks are getting deployed in real world, we need to make sure that the predictions made by those models are reliable.
- It was observed that the shallow networks were well calibrated but the modern deep neural networks albeit very powerful in learning myriad of complex probles tends to be ill-calibrated.
- This becomes also importtant for unceratinty estimation as we treat the predicted probability as our confidence anf take decision around this probability score , we need to be sure about this, say for instance in medical domain when the predcited score of model is 80% you don't condisder visiting doctor but if its below 80 you consider visiting docto and
