```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bin_conf = [0,0.17,0.29,0.5067,0.7075,0.9067,1]
bin_acc = [0,0.5,0.667,0.333,0.5,0.667,1]
```

```python
plt.figure()
plt.plot([0,1], [0,1], '--', color='tab:gray', label = 'ideal')
plt.plot(bin_conf,bin_acc, '-o', color='tab:purple', label='NN')
plt.plot(bin_conf,bin_acc, '-o', color='tab:purple', label='NN')
plt.xlabel('Predicted Cumulative Distribution')
plt.ylabel('Empirical cumulative distribution')
plt.xlabel('Predicted Confidence')
plt.ylabel('Observed Confidence (accuracy)')
plt.legend()
plt.title('Calibration Plot')
plt.show()
```
